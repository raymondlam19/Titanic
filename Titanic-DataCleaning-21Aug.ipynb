## Data Cleaning

```python
# df_train
# null: age, cabin, 'embarked'
# create new feature: title, family size, IsAlone
# delete: PassengerId, name, ticket, cabin (too many null), SibSp, Parch
# convert to categories/one-hot: title, sex, Embarked

# df_test
# null: age, cabin, 'fare'
# create new feature: title, family size, IsAlone
# delete: PassengerId, name, ticket, cabin (too many null), SibSp, Parch
# convert to categories/one-hot: title, sex, Embarked
```

### 1) fillna

```python
df_train['Embarked'].unique()
```

```python
df_train['Embarked'].value_counts()
```

```python
df_train['Embarked'].mode()[0]
```

```python
#df_train
#null: age, cabin, 'embarked'

#df_test
#null: age, cabin, 'fare'

for df in df_cleaner:
    # fill with median
    df['Age'] = df['Age'].fillna(df['Age'].median())
    
    # fill na
    df['Cabin'] = df['Cabin'].fillna("")

df_test['Fare'] = df_test['Fare'].fillna(df_test['Fare'].median())
df_train['Embarked'] = df_train['Embarked'].fillna(df_train['Embarked'].mode()[0])
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

### 2) Create new feature

```python
aaa = df_train['Name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]
aaa[0]
```

```python
#df_train
#create new feature: title, family size, IsAlone

#df_test
#create new feature: title, family size, IsAlone

for df in df_cleaner:
    # create new features
    df['Title'] = df['Name'].str.split(", ", expand=True)[1].str.split(".", expand=True)[0]
    df['Family_size'] = df['Parch'] + df['SibSp'] + 1
    df['IsAlone'] = np.where((df['SibSp'] == 0) & (df['Parch'] == 0), 1, 0)
```

```python
df_train['Title'].unique()
```

```python
df_train['Title'].value_counts()
```

```python
df_test['Title'].unique()
```

```python
df_test['Title'].value_counts()
```

```python
#cleanup rare titles (<7)

stat_min = 7
for df in df_cleaner:
    other_title = (df['Title'].value_counts() < stat_min)
    df['Title'] = df['Title'].apply(lambda x: 'Other' if other_title.loc[x] == True else x)
```

```python
df_train['Title'].value_counts()
```

```python
df_test['Title'].value_counts()
```

### 3) Delete features

```python
# df_train
# delete: PassengerId, name, ticket, cabin (too many null), SibSp, Parch

# df_test
# delete: PassengerId, name, ticket, cabin (too many null), SibSp, Parch

drop_column = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp', 'Parch']
for df in df_cleaner:
    df.drop(drop_column, axis=1, inplace = True)
```
